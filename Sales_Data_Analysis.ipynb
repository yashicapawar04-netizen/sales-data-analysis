```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("sales_data.csv")
df.head()
```

```python
df["Total_Sales"]=df["Quantity"]*df["Price"]
df.head()
```

```python
df["Total_Sales"].sum()
```

```python
df.groupby("Product")["Quantity"].sum().sort_values(ascending=False)
```

```python
df.groupby("Region")["Total_Sales"].sum()
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df["Month"]=df["Date"].dt.month
df.groupby("Month")["Total_Sales"].sum()
```

```python
df.groupby("Product")["Total_Sales"].sum().plot(kind="bar")
plt.title("Product-wise Sales")
plt.show()
```

identify best-selling products, and understand regional and monthly sales trends.
